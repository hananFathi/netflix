### Distribution of 'type' (movies vs Tv shows)

```python
type_distribution = shows['type'].value_counts(normalize=True)*100
plt.figure(figsize=(6,4))
type_distribution.plot(kind='bar',color=['blue','orange'])
plt.title('Distribution of Content Type')
plt.xlabel('Type')
plt.ylabel('Percentage')
plt.tight_layout()
plt.show()
```

### Distribution of Year 

```python
plt.figure(figsize=(10,6))
shows['release_year'].hist(bins=30,color='skyblue',edgecolor='black')
plt.title('Distribution of Release Years')
plt.xlabel('Release Year')
plt.ylabel('Frequency')
plt.show()
```

### Distribution of Rating

```python
rating_distribution = shows['rating'].value_counts(normalize=True)*100
plt.figure(figsize=(10,6))
rating_distribution.plot(kind='bar',color='green')
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Percentage')
plt.show()
```

### Distribution of Content by Country

```python
country_distribution = shows['country'].value_counts().head(10)  # Top 10 countries
plt.figure(figsize=(12, 6))
country_distribution.plot(kind='bar', color='skyblue')
plt.title('Top 10 Countries Contributing Content')
plt.xlabel('Country')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Distribution of Content duration for Movies and Tv Shows

```python
# Separate movies and TV shows
movies = shows[shows['type'] == 'Movie']
tv_shows = shows[shows['type'] == 'TV Show']
movies
```

```python
# Explore the distribution of content duration
plt.figure(figsize=(12, 6))
plt.hist(movies['duration'].head(50), bins=30, color='skyblue', alpha=0.7, label='Movies')
plt.hist(tv_shows['duration'].head(50), bins=30, color='orange', alpha=0.7, label='TV Shows')
plt.title('Distribution of Content Duration')
plt.xlabel('Duration (Minutes for Movies, Seasons for TV Shows)')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.legend()
plt.grid(False)
plt.tight_layout()
plt.show()
```

```python
# Explore the distribution of content across genres
plt.figure(figsize=(12, 8))
genre_distribution = shows['listed_in'].str.split(', ').explode().value_counts().head(15)
sns.barplot(y=genre_distribution.index, x=genre_distribution.values, palette='viridis')
plt.title('Distribution of Content Across Genres')
plt.xlabel('Number of Titles')
plt.ylabel('Genre')
plt.tight_layout()
plt.show()
```

### Relation between Release Year and Rating

```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=shows, x='release_year', y='rating')
plt.title('Relationship Between Release Year and Rating')
plt.xlabel('Release Year')
plt.ylabel('Rating')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
